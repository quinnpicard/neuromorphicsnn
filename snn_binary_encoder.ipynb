```python
#may need to install some of these libraries 
import librosa
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Loading an audio file which is saved in the same place as this script 
audio_file = "song2.wav"
audio_signal, sample_rate = librosa.load(audio_file, sr=None) # we may want to hardcode our sampling rate if it is not consistant between songs 
```

```python
# Computing the Short-Time Fourier Transform (STFT) magnitude spectrum
'''
these constants allow us to prioritize frequency or time reolution, where larger n_fft values provide increases in frequency resolution,
and larger hop_length values 

'''

n_fft = 4096 #determines the length of each frame
hop_length = n_fft // 4 #number of audio samples between successive frames in the Short-Time Fourier Transform (STFT) analysis
magnitude_spectrum = np.abs(librosa.stft(audio_signal, n_fft=n_fft, hop_length=hop_length))

# Appling Mel filterbank from librosa 
num_mels = 20 #128 originally, but likely too much computation time and no longer actually follows any biological model
mel_spectrum = librosa.feature.melspectrogram(
    sr=sample_rate,
    S=magnitude_spectrum,
    n_fft=n_fft,
    hop_length=hop_length,
    n_mels=num_mels
)
```

```python
pd.DataFrame(mel_spectrum)
```

```python
def plot_mel_spectrogram(mel_spectrum, sample_rate, hop_length):
    # Convert Mel spectrogram to dB scale
    mel_spectrogram_db = librosa.power_to_db(mel_spectrum, ref=np.max)

    # Compute the time axis in seconds
    time = np.arange(mel_spectrogram_db.shape[1]) * hop_length / sample_rate

    # Plot the Mel spectrogram
    plt.figure(figsize=(10, 6))
    plt.imshow(mel_spectrogram_db, origin='lower', aspect='auto', cmap='viridis',
               extent=[0, time[-1], 0, mel_spectrum.shape[0]])
    plt.colorbar(format='%+2.0f dB')
    plt.title('Mel Spectrogram')
    plt.xlabel('Time (s)')
    plt.ylabel('Mel Frequency')
    plt.tight_layout()
    plt.show()
plot_mel_spectrogram(mel_spectrum, sample_rate = sample_rate, hop_length = 512)
```

```python
'''
pros and cons of this implementation for reducing to binary data

pros:
- fairly computationally efficient as it does not require loops or nested conditional statements 

cons:
- makes all channels have the same number of 1s so all channels are equivelently valued 
- does not care about locality in time or frequency space 
- in relatively inactive channels it will still force the same number of 1 outputs leading 
  to fake spikes  

'''

def binary_mel_spectrogram(mel_spectrum):
    # Compute the row-wise mean values (the mean of each channel)
    channel_means = np.mean(mel_spectrum, axis=1)

    # Apply thresholding to create binary representation
    binary_spectrum = np.where(mel_spectrum >= channel_means[:, np.newaxis], 1, 0)

    return binary_spectrum

binary_spectrum = binary_mel_spectrogram(mel_spectrum)
pd.DataFrame(binary_spectrum)
```

```python
# repeated pattern is the chorus which appears at about t = 100s and t = 200s 
plot_mel_spectrogram(binary_spectrum, sample_rate = sample_rate, hop_length = 512)
```
